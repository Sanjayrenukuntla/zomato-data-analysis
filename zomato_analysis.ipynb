2 .Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#for better visuals
sns.set(style='whitegrid')
```

3 .load Dataset

```python
df = pd.read_csv("Zomato.csv")

#Clean Column Names
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace("(", "")
df.columns = df.columns.str.replace(")", "")

df.head()
```

4 .Basic Exploration

```python
# print('shape:',df.shape)
# df.info()
# df.describe()
# df.isnull().sum()
```

5 .Data cleaning

```python
#remove missing values
df.dropna(inplace = True)
#Remove Duplicates
df.drop_duplicates(inplace = True)
```

6 .Ratings Distribution

```python
# Analyze rating distribution to understand customer satisfaction
plt.figure(figsize=(8,5))

sns.histplot(df['rate'], bins=10)

plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')

plt.show()

# Insight:
# Most restaurants have ratings between 3.5 and 4.0.
```

7 .Cost Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df['rate'], bins=10)

plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')

plt.show()

# Insight:
# Most restaurants have ratings between 3.5 and 4.0.
```

8 .Online Order vs Ratings

```python
plt.figure(figsize=(6,4))

sns.boxplot(x='online_order', y='rate', data=df)

plt.title('Online Order vs Rating')

plt.show()

# Insight:
# Restaurants offering online orders have slightly higher ratings.
```

9.Table Booking vs Ratings

```python
plt.figure(figsize=(6,4))

sns.boxplot(x='book_table', y='rate', data=df)

plt.title('Table Booking vs Rating')

plt.show()

# Insight:
# Restaurants with table booking tend to have higher ratings.
```

10.Correlation Heatmap

```python
plt.figure(figsize=(6,4))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")

plt.show()

# Insight:
# There is a positive relationship between cost and votes.
```
