2 .Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#for better visuals
sns.set(style='whitegrid')
```

5 .Data cleaning

```python
#remove missing values
df.dropna(inplace = True)
#Remove Duplicates
df.drop_duplicates(inplace = True)
```

6 .Ratings Distribution

```python
## Analyze rating distribution to understand customer satisfaction
plt.figure(figsize = (15,5))
sns.histplot(df['rate'],bins = 5,kde=True)
plt.title('Distribution Of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()

# Insight:
# Most restaurants hit the 3.5-4.0 range. It's a solid baseline, but the real competitive edge lies in breaking into the 4.5+ elite tier, which remains sparse.
```

7 .Cost Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['approx_cost(for two people)'],bins = 10)
plt.title('Cost Distribution')
plt.xlabel('Cost')
plt.ylabel('Count')
plt.show()

# Insight:
# The market clearly favors mid-range pricing. There's a massive volume of casual diners, while the ultra-luxury segment remains a niche, low-volume play.
```

8 .Online Order vs Ratings

```python
#Compare impact of online ordering on ratings
plt.figure(figsize = (6,4))
sns.boxplot(x='online_order',y='rate',data=df)
plt.title("Online Order vs Ratings")
plt.show()

# Insight:
# Online ordering doesn't just add convenience; it shifts the rating floor. Delivery-optimized spots seem to maintain a tighter consistency in customer feedback.
```

9.Table Booking vs Ratings

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='book_table',y='rate',data=df)
plt.title("Table Booking vs Ratings")
plt.show()

# Insight:
# Booking availability is a strong proxy for quality. Places that offer reservations generally command better ratings, likely due to a more curated dining experience.
```

10.Correlation Heatmap

```python
plt.figure(figsize=(6,4))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Insight:
# Higher price points correlate with higher vote counts. Essentially, premium venues are better at driving engagement and 'buzz' than budget alternatives.
```

11.Top Restaurant Types

```python
top_types = df["listed_in(type)"].value_counts().head(10)
display(top_types)

# Insight:
# Dining and Cafes dominate the landscape. For a new entrant, this confirms a proven market but also highlights high saturation in these specific categories.
```
